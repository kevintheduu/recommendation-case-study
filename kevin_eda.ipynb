```python
import pandas as pd

def get_lines():
    with open('data/movies.dat', 'rb') as f:
        for line in f: 
            yield line

def get_weird_lines():
    for i, line in enumerate(get_lines()):
        for char in line: 
            if char > 127:
                yield line
```
