```python
import pandas as pd
import numpy as np
```

```python
import pyspark

spark = (pyspark.sql.SparkSession.builder 
  .master("local[*]")
  .getOrCreate())

meta_df = spark.read.csv('data/movies_metadata.csv',
                               inferSchema=True,
                               header=True
                               ).toPandas()
```

```python
movies_df = pd.read_csv('data/movies.dat',
            header=None, sep='::', engine='python')
```

```python
movies_df.head()
```

```python
movies_df.columns=['id','title','category']
```

```python
#movies_df.loc[:,'year'] = movies_df.title.map(lambda x: x[-5:-1])
```

```python
movies_df.head()
```

```python
movies_df['title'] = movies_df['title'].map(lambda x: x.rpartition(' (')[0])
```

```python
movies_df.head()
```

```python
movies_df['category'] = movies_df['category'].map(lambda x: x.split('|'))
```

```python
movies_df.head()
```

```python
meta_df.columns
```

```python
meta_df = meta_df.drop(index=meta_df[meta_df.release_date.isna()].index)
```

```python
meta_df.loc[:,'year'] = meta_df.release_date.map(lambda x: x[:4])

# meta_df[meta_df['year'].contains]
```

```python
meta_df_clean_years = meta_df.drop(index=meta_df[meta_df.year.map(
    lambda x: not (x.startswith('19') or x.startswith('20')))].index)
```

```python
meta_df_clean_years.year.astype(float)
```

```python
meta_df_clean_years.year.unique()
```

```python
movies_df['title_year'] = list(zip(movies_df.title, movies_df.year))
```

```python
meta_df_clean_years['title_year'] = list(zip(meta_df_clean_years.title,
                                             meta_df_clean_years.year))
```

```python
meta_df_clean_years.head()
```

```python
movies_df.head()
```

```python
results = pd.merge(movies_df,
                  meta_df[['title','year','budget','revenue', 
                     'original_title',
                           'runtime','vote_average','title_year','status',
                     'popularity','vote_count']],
                  left_on='title_year',
                  right_on='title_year',
                  how='left')
```

```python
results.columns
```

```python
features = results[['id', 'title_x', 'category', 'year_x', 'title_year',
       'budget', 'revenue', 'runtime',
       'vote_average', 'popularity', 'vote_count']]
```

```python
features = features.rename(columns={'title_x':'title','year_x':'year'})
```

```python
features.head()
```

```python
features.to_csv('features.csv',index=False)
```
