```python
import pandas as pd
import numpy as np
```

```python
import pyspark

spark = (pyspark.sql.SparkSession.builder 
  .master("local[*]")
  .getOrCreate())

meta_df = spark.read.csv('data/movies_metadata.csv',
                               inferSchema=True,
                               header=True
                               ).toPandas()
```

```python
movies_df = pd.read_csv('data/movies.dat',
            header=None, sep='::', engine='python')
```

```python
movies_df.head()
```

```python
type(movies_df)
```

```python
movies_df[1] = movies_df[1].map(lambda x: x.split(' (')[0])
```

```python
movies_df.head()
```

```python
movies_df[2] = movies_df[2].map(lambda x: x.split('|'))
```

```python
movies_df.head()
```

```python
meta_df.columns
```

```python
movies_df.columns=['id','title','category']
```

```python
movies_df.category
```

```python
meta_df.shape
```

```python
results = pd.merge(meta_df,
                  movies_df[['id','title','category']],
                  left_on='original_title',
                  right_on='title',
                  how='left')
```

```python
features = (results[['id_y','title_y','category','budget','revenue', 'original_title',
                           'runtime','vote_average','status','popularity','vote_count']])
```

```python
features = features.rename(columns={'id_y':'id'})
```

```python
features.shape
```

```python
meta_df.shape
```

```python
features.to_csv('features.csv',index=False)
```
